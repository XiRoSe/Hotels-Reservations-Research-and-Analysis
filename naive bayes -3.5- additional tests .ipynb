```python
import pandas as pd
from datetime import datetime
import numpy as np

df = pd.read_csv("Hotels_Good_Data.csv",parse_dates=[1,2])
df = df.drop(['Unnamed: 0','Discount Price'],axis=1)

minSnap_date = min(df['Snapshot Date'])
minCheck_date = min(df['Checkin Date'])

df['Snapshot Date']= (df['Snapshot Date']-minSnap_date).dt.days
df['Checkin Date'] = (df['Checkin Date']-minCheck_date).dt.days

gb_df = df.groupby(['Snapshot Date','Checkin Date','Hotel Name','Day Diff','WeekDay'])
gb_df = df.loc[gb_df['Discount Perc'].idxmax()]
gb_df = gb_df.drop(['Discount Perc','Discount Diff'], axis=1)
lookup = np.sort(np.unique(gb_df['Hotel Name']))
gb_df['Hotel Name'] = np.searchsorted(lookup,gb_df['Hotel Name'])
```

```python
import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
from sklearn.naive_bayes import GaussianNB  
from sklearn.cross_validation import train_test_split
gnb = GaussianNB()
X=gb_df.drop(['Discount Code','Snapshot Date','Checkin Date','Hotel Name','Day Diff'],axis=1)
Y=gb_df['Discount Code']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=1)

gnb.fit(X_train,Y_train)
```

```python
from sklearn import metrics

Y_pred = gnb.predict(X_test)
print('The Accuracy for this model is -  ',metrics.accuracy_score(Y_test,Y_pred))
```

```python
def OneVsAll(x):
    counts = list([0,0,0,0,0])
    maximum = 0
    pred1 = gnb1.predict(x)
    pred2 = gnb2.predict(x)
    pred3 = gnb3.predict(x)
    pred4 = gnb4.predict(x)
    y_pred=list()
   # print(pred2)
    for i in range(len(x)):
        if(pred1[i] == 1):
            #print('true 1')
            counts[1] += 1
        else:
            #print('false 1')
            counts[2] += 1
            counts[3] += 1
            counts[4] += 1
            
        if( pred2[i] == 2):
           # print('true 2')
            counts[2] += 1
        else:
           # print('false 2')
            counts[1] += 1
            counts[3] += 1
            counts[4] += 1
            
        if( pred3[i] == 3):
           # print('true 3')
            counts[3] += 1
        else:
            #print('false 3')
            counts[1] += 1
            counts[2] += 1
            counts[4] += 1
            
        if( pred4[i] == 4):
            #print('true 4')
            counts[4] += 1
        else:
           # print('false 4')
            counts[1] += 1
            counts[2] += 1
            counts[3] += 1
        
       # print(counts)
        
        for index, c in enumerate(counts):
            if (c > maximum):
                maximum = c
                idx = index
       
        
        #for index,count in enumerate(counts):
         #   if (count == maximum):
         #       maxCount.append(index)
          #  if (len(maxCount)==2):
                
            
        y_pred.append([idx,counts])
        counts=list([0,0,0,0,0])
        maximum=0        
        
        
    return y_pred
```

```python
def lern(x):
    y_pred=list()
    ovsa=OneVsAll(x)
    ovso=OneVsOne(x)
    for i in range(len(ovsa)):
        if (ovsa[i][0]==ovso[i][0]):
            y_pred.append(ovsa[i][0])
        else:
            if (ovsa[i][1][ovsa[i][0]]>ovso[i][1][ovso[i][0]]):
                y_pred.append(ovsa[i][0])
            else:
                y_pred.append(ovso[i][0])
    return y_pred
```
